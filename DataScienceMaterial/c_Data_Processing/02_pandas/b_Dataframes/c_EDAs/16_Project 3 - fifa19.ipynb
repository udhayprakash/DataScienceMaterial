```python
# https://www.kaggle.com/kushal1996/fifa-19-analysis-dream-team
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
fifa19 = pd.read_csv("../../datasets/fifa19.csv")
fifa19.head(3)
```

```python
# drop column 'Unnamed: 0'
fifa19 = fifa19.drop("Unnamed: 0", 1)
fifa19.columns
```

```python
fifa19.head(3)
fifa19.Age.describe()
```

# Age

```python
# 5 oldest players
fifa19.sort_values(by=["Age"], ascending=False, inplace=True)
fifa19.head(5)
```

```python
# 5 youngest players
fifa19.sort_values(by=["Age"], ascending=True, inplace=True)
fifa19.head(5)
```

```python
# mean of ages
fifa19.Age.mean()
```

```python
# draw histogram
fifa19.Age.plot(kind="hist")
```

```python
# describe age
fifa19.Age.describe()
```

# Player by Country

```python
top10 = (
    fifa19.groupby("Nationality")
    .count()
    .sort_values(by=["Name"], ascending=False)
    .head(10)
)
top10
```

```python
# show up top 10 on bar chart
top10["ID"].plot(kind="barh")
```

# Player by Club

```python
fifa19.groupby("Club").count().sort_values(by="ID", ascending=False).head()
```

# Top Player by Value

```python
fifa19.Value.describe()
```

```python
# check unique value
fifa19.Value.unique()
```

```python
# convert value data
def convert_value(value):
    if value.endswith("K"):
        return float(value.replace("€", "").replace("K", "")) * 1000
    if value.endswith("M"):
        return float(value.replace("€", "").replace("M", "")) * 1000000


fifa19.Value = fifa19.Value.apply(convert_value)
fifa19.head()
```
