```python
# https://www.kaggle.com/rounakbanik/ted-data-analysis
import pandas as pd

df = pd.read_csv("../../datasets/ted_main.csv")
df.head(3)
```

```python
# print out the columns
df.columns
```

```python
# reorder the columns
df = df[
    [
        "name",
        "title",
        "description",
        "main_speaker",
        "speaker_occupation",
        "num_speaker",
        "duration",
        "event",
        "film_date",
        "published_date",
        "comments",
        "tags",
        "languages",
        "ratings",
        "related_talks",
        "url",
        "views",
    ]
]
df.head(3)
```

```python
# change from epoc time to human readable datetime
import datetime

df["film_date"] = df["film_date"].apply(
    lambda x: datetime.datetime.fromtimestamp(int(x)).strftime("%d-%m-%Y")
)
df["film_date"].head(3)
```

```python
df["published_date"] = df["published_date"].apply(
    lambda x: datetime.datetime.fromtimestamp(int(x)).strftime("%d-%m-%Y")
)
df["published_date"].head(3)
```

```python
df.head(3)
```

```python
# how many row in this dataframe
df.shape
```

```python
# what is 5 most viewed talks of all time
pop_talks = df.sort_values("views", ascending=False)[["title", "main_speaker", "views"]]
pop_talks.head(5)
```

```python
%matplotlib inline
import seaborn as sns

sns.distplot(df["comments"])
```

```python
# analyse ted talk on month and year
# separate the month column
month_order = [
    "Jan",
    "Feb",
    "Mar",
    "Apr",
    "May",
    "Jun",
    "Jul",
    "Aug",
    "Sep",
    "Oct",
    "Nov",
    "Dec",
]
df["month"] = df["film_date"].apply(lambda x: month_order[int(x.split("-")[1]) - 1])
df["month"].head(10)
```

```python
month_df = pd.DataFrame(df["month"].value_counts()).reset_index()
month_df.columns = ["month", "talks"]
month_df.head(3)
```

```python
sns.barplot(x="month", y="talks", data=month_df, order=month_order)
```

```python
df["year"] = df["film_date"].apply(lambda x: x.split("-")[2])
df["year"].head(3)
```

```python
year_df = pd.DataFrame(df["year"].value_counts().reset_index())
year_df.columns = ["year", "talks"]
year_df.head(3)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 5))
sns.pointplot(x="year", y="talks", data=year_df)
```
