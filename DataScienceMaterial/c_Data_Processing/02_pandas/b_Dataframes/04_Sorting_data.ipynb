#### Loading the libraries

```python
import pandas as pd
from pathlib import Path

DATASETS_PATH = Path("../../../../datasets/")
```

```python
df_imdb = pd.read_csv(Path(DATASETS_PATH / "IMDB_Movie_Sorting.csv"))
df_imdb.head(3)
```

# Ten hightest score movie

```python
df_imdb.sort_values("imdb_score", ascending=False).head(10)
```

```python
df_imdb.head()  # original df is not changed
```

```python
df_imdb[["movie_title", "imdb_score", "title_year"]].head()
```

```python
# explain inplace parameter
# inplace will make change right inside dataframe so do not need another frame to store data after sort
df_imdb.sort_values(by=["imdb_score"], ascending=False, inplace=True)
```

```python
df_imdb[["movie_title", "imdb_score", "title_year"]].head()  # original df is changed
```

```python
# explain ascending parameter
# default value for ascending is True
# sorted with a list of column
df_imdb.sort_values(by=["imdb_score", "title_year"], ascending=False).head(10)
```

```python
df_imdb[["movie_title", "imdb_score", "title_year"]].head()
```
