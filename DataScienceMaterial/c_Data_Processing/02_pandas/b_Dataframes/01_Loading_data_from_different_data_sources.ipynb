### 6. loading from a python class instance

```python
class Company:
    def __init__(self, name, addr, comp_val):
        self.name = name
        self.addr = addr
        self.__comp_val = comp_val

    def to_dict(self):
        return {"name": self.name, "addr": self.addr, "comp_val": self.__comp_val}


# instantiation
c1 = Company("apple", "addr1", 1111)
c2 = Company("microsoft", "addr2", 2222)
c3 = Company("tesla", "addr3", 3333)

df = pd.DataFrame([inst.to_dict() for inst in [c1, c2, c3]])
print(df)
```

### 8. Reading from tabular data file - csv, excel, tsv, .. table-like data format

read_table assumptions:
- file is separated by tabs
- presence of a header role

```python
# reading a well-formatted .tsv file
url = "http://bit.ly/chiporders"
orders = pd.read_table(url)

orders.head()
```

```python
url2 = "http://bit.ly/movieusers"
users = pd.read_table(url2)

users.head()
```

```python
user_cols = ["user_id", "age", "gender", "occupation", "zip_code"]
users = pd.read_table(url2, sep="|", header=None, names=user_cols)

users.head()
```
